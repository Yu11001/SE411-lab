```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
```

```python
data = pd.read_csv('LAB08_sat_gpa.csv')
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

1. SAT coef. stands for?
2. Constant coef. stands for?
3. R-squared is?

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat, lw=4, c='orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
data = pd.read_csv('cleaned_all_phones.csv')
data.describe()
```

```python
y = data['price(USD)']
x1 = data['inches']
plt.scatter(x1,y)
plt.xlabel('inches', fontsize = 20)
plt.ylabel('price(USD)', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 107.7878*x1 + (-354.4156)
fig = plt.plot(x1,yhat, lw=4, c='orange', label = 'regression line')
plt.xlabel('inches', fontsize = 20)
plt.ylabel('price(USD)', fontsize = 20)
plt.show()
```
